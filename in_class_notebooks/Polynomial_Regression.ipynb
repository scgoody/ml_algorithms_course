# Polynomial Regression

## Data Preprocessing

dataset: https://www.kaggle.com/datasets/mariospirito/position-salariescsv

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
datapath = "C:/Users/scgoo/OneDrive/Documents/colleges/Jacksonville/Spring 2025/Math 470 - ML Algorithms/githubMath470/data/"
dataset = pd.read_csv(datapath+"Position_Salaries.csv")
dataset
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
plt.scatter(X,y)
plt.title("Position VS Salary")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

## Fitting Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
plt.scatter(X,y, color="red")
plt.plot(X,lin_reg.predict(X),color="blue")
plt.title("Linear Regression - Position VS Salary")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
lin_reg.predict([[6.5]])
```

```python
accuracy = lin_reg.score(X,y)
print('Accuracy of the model is ', accuracy*100, '%')
```

## Fitting Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
X_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
plt.scatter(X,y, color="red")
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color="blue")
plt.title("Polynomial Regression - Position VS Salary")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```

```python
accuracy = lin_reg2.score(poly_reg.fit_transform(X),y)
print('Accuracy of the model is ', accuracy*100, '%')
```
